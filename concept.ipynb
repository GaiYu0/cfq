```python
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
```

```python
data_dir = '/work/yu_gai/cfq'
```

```python
d = np.load(f'{data_dir}/concept.npz')
con = d['con']
ncon = d['n_con']
indptr = np.cumsum(np.hstack([[0], ncon]))
```

```python
def plot(split):
    d = np.load(f'{data_dir}/splits/{split}.npz')
    train_idx, dev_idx, test_idx = d['trainIdxs'], d['devIdxs'], d['testIdxs']

    get = lambda idx: np.hstack([con[indptr[i] : indptr[i + 1]] for i in idx])
    
    def plot_(con):
        _, counts = np.unique(con, return_counts=True)
        pl.plot(np.arange(len(counts)), counts / counts.sum(), '_')

    plot_(con)
    plot_(get(train_idx))
    plot_(get(dev_idx))
    plot_(get(test_idx))
```
