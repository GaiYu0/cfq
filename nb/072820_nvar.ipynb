```python
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
```

```python
data_dir = '/work/yu_gai/cfq'
```

```python
nvar = np.load(f'{data_dir}/nvar.npz')['n_var']
```

```python
def plot(split):
    d = np.load(f'{data_dir}/splits/{split}.npz')
    train_idx, dev_idx, test_idx = d['trainIdxs'], d['devIdxs'], d['testIdxs']

    def plot_(nvar):
        _, counts = np.unique(nvar, return_counts=True)
        pl.plot(np.arange(len(counts)), counts / counts.sum(), '_')

    plot_(nvar)
    plot_(nvar[train_idx])
    plot_(nvar[dev_idx])
    plot_(nvar[test_idx])
```
